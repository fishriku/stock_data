```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# 
date_range = pd.date_range(start='1/1/2020', end='1/1/2023')

# 
stock_prices = [random.randint(100, 500) for _ in range(len(date_range))]

# 
df = pd.DataFrame(data={'Date': date_range, 'Price': stock_prices})
df.set_index('Date', inplace=True)

# 
plt.figure(figsize=(10,5))
plt.plot(df.index, df['Price'])
plt.title('Random Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
plt.show()
```
